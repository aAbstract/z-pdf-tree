```python
import fitz
import matplotlib.pyplot as plt
```

```python
file_path = '../data/sample_rxi_12.pdf'
doc = fitz.open(file_path)
toc_links = []
for page in doc:
    link = page.first_link
    while link:
        link_text = page.get_textbox(link.rect)
        toc_links.append(link_text)
        link = link.next
```

```python
len_dist = [len(x) for x in toc_links]
plt.hist(len_dist)
```

```python
# find short link threshold
[x for x in toc_links if len(x) > 100 and len(x) < 200]
```
